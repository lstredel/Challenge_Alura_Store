### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
```

```python
tienda1.insert(loc=0, column='Tienda Origen', value="Tienda 1")
tienda2.insert(loc=0, column='Tienda Origen', value="Tienda 2")
tienda3.insert(loc=0, column='Tienda Origen', value="Tienda 3")
tienda4.insert(loc=0, column='Tienda Origen', value="Tienda 4")
```

#1. Análisis de facturación



```python
pd.options.display.float_format = '{:,.0f}'.format
```

```python
ventas_totales_por_tienda = Tiendas_Concatenadas.groupby("Tienda Origen")["Precio"].agg(["sum","count"])
ventas_totales_por_tienda.head()
```

# 2. Ventas por categoría

```python
ventas_totales_por_categoria = Tiendas_Concatenadas.groupby(["Categoría del Producto"])["Precio"].agg(["sum", "count"])
display(ventas_totales_por_categoria.head(10))
```

```python
ventas_totales_por_tienda_y_categoria = Tiendas_Concatenadas.groupby(["Tienda Origen", "Categoría del Producto"])["Precio"].agg(["sum", "count"])
display(ventas_totales_por_tienda_y_categoria.head(50))
```

```python
ventas_totales_por_tienda_y_categoria = Tiendas_Concatenadas.groupby(["Categoría del Producto","Tienda Origen"])["Precio"].agg(["sum", "count"])
display(ventas_totales_por_tienda_y_categoria.head(50))
```

# 3. Calificación promedio de la tienda


```python
Calificacion_Tiendas = Tiendas_Concatenadas.groupby(["Tienda Origen", "Calificación"])["Calificación"].count().rename("Cantidad")
Calificacion_Tiendas.head(30)

```

```python
total_por_tienda = Tiendas_Concatenadas.groupby("Tienda Origen")["Calificación"].count().rename("Total Calificado")
total_por_tienda.head()
```

```python
df_frecuencias = Calificacion_Tiendas.reset_index()
```

```python
df_totales = total_por_tienda.reset_index()
```

```python
df_combinado = df_frecuencias.merge(
    df_totales[['Tienda Origen', 'Total Calificado']],
    on='Tienda Origen',
    how='left'
)
df_combinado.head(30)
```

```python
df_combinado['Porcentaje por Calificación'] = (
    df_combinado['Cantidad'] / df_combinado['Total Calificado']
) * 100
pd.options.display.float_format = '{:,.0f}'.format
df_combinado.head(30)
```

# 4. Productos más y menos vendidos

```python
df_ventas_por_producto = Tiendas_Concatenadas.groupby('Producto').size().reset_index(name='Total_Cantidad_Vendida_Global')
```

```python
N = 10
df_productos_top_10 = df_ventas_por_producto.sort_values(
    by='Total_Cantidad_Vendida_Global',  # Columna de conteo
    ascending=False                     # De mayor a menor
).head(N)
print(f"--- Top {N} Productos Más Vendidos (por Conteo) ---")
display(df_productos_top_10)
```

```python
N = 10
df_productos_bottom_10 = df_ventas_por_producto.sort_values(
    by='Total_Cantidad_Vendida_Global',  # Columna de conteo
    ascending=True                     # De menor a mayor
).head(N)
print(f"--- Top {N} Productos Menos Vendidos (por Conteo) ---")
display(df_productos_bottom_10)
```

```python
productos_top_global = df_productos_top_10['Producto'].unique()
df_solo_top_productos = Tiendas_Concatenadas[
    Tiendas_Concatenadas['Producto'].isin(productos_top_global)
]
df_participacion_tienda = df_solo_top_productos.groupby(['Producto', 'Tienda Origen']).size().reset_index(name='Cantidad_Vendida_Tienda')
df_participacion_final = df_participacion_tienda.merge(
    df_ventas_por_producto,
    on='Producto',
    how='left'
)
df_participacion_final['Porcentaje_Participacion'] = (
    df_participacion_final['Cantidad_Vendida_Tienda'] / df_participacion_final['Total_Cantidad_Vendida_Global']
) * 100
pd.options.display.float_format = '{:,.0f}'.format

resumen_styled = df_participacion_final.style.format({
    'Cantidad_Vendida_Tienda': '{:,.0f}',
    'Total_Cantidad_Vendida_Global': '{:,.0f}',
    'Porcentaje_Participacion': '{:,.2f}%' # Porcentaje con 2 decimales
})

print("--- Porcentaje de Participación por Tienda (Solo Productos Top 10) ---")
display(resumen_styled)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 1. Preparar los datos del DataFrame
# 'ventas_totales_por_tienda' es una Serie, por lo que sort_values() se aplica directamente a sus valores.
df_plot = ventas_totales_por_tienda.sort_values(ascending=False)

# Etiquetas (nombres de las tiendas) se obtienen del índice de la Serie
etiquetas = df_plot.index

# Valores (montos de ventas) se obtienen de los valores de la Serie
valores = df_plot.values

# 2. Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(10, 8))

# 3. Dibujar el gráfico circular
ax.pie(
    valores,
    labels=etiquetas,
    autopct='%1.1f%%',  # Mostrar porcentaje con 1 decimal
    startangle=90,      # Iniciar el primer sector en la parte superior
    shadow=True,
    wedgeprops={'edgecolor': 'black'} # Añade un borde para mayor claridad
)

# 4. Ajustar y añadir título
ax.axis('equal')  # Asegura que el gráfico sea un círculo perfecto
plt.title('Distribución Porcentual de Ventas por Tienda', fontsize=16)

# 5. Mostrar el gráfico
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 7))

# 3. Dibujamos el gráfico de barras agrupadas usando Seaborn
# x: La combinación de Tienda y Producto. Usaremos el índice para plotear.
# y: El valor de la cantidad vendida.
# hue: Usamos 'Tienda Origen' para diferenciar visualmente las barras por tienda.

sns.barplot(
    data=df_participacion_final, # Changed data from 'top_10_definitivo' to 'df_participacion_final' as 'top_10_definitivo' is not defined in the kernel.
    x='Producto',
    y='Cantidad_Vendida_Tienda', # Changed y-axis to 'Cantidad_Vendida_Tienda' from 'Total_Cantidad_Vendida_Producto_Top' as it's available in 'df_participacion_final'
    hue='Tienda Origen',
    palette='viridis' # Esquema de colores
)

# 4. Ajustar etiquetas y título
plt.title('Top 10 Cantidades Vendidas (Producto y Tienda)', fontsize=16)
plt.xlabel('Producto', fontsize=12)
plt.ylabel('Cantidad Vendida', fontsize=12)
plt.xticks(rotation=45, ha='right') # Rotar las etiquetas del eje X para que no se superpongan
plt.legend(title='Tienda')
plt.grid(axis='y', linestyle='--') # Añadir líneas de cuadrícula para facilitar la lectura

# 5. Mostrar el gráfico
plt.tight_layout() # Ajusta automáticamente los parámetros de la figura para que no se superpongan
plt.show()
```

```python
resumen_por_tienda = Tiendas_Concatenadas.groupby('Tienda Origen')['Calificación'].mean().reset_index(name='Promedio General de Calificaciones')

# 1. Preparar los datos
# Si quieres ordenar el gráfico por la calificación promedio (de mayor a menor)
df_plot = resumen_por_tienda.sort_values(
    by='Promedio General de Calificaciones',
    ascending=False
).reset_index(drop=True)

# 2. Crear la figura y los ejes
plt.figure(figsize=(10, 6))

# 3. Dibujar el gráfico de barras con Seaborn
sns.barplot(
    data=df_plot,
    x='Tienda Origen',
    y='Promedio General de Calificaciones',
    palette='Blues_d' # Un esquema de color agradable
)

# 4. Ajustar etiquetas y título
plt.title('Promedio General de Calificaciones por Tienda', fontsize=16)
plt.xlabel('Tienda Origen', fontsize=12)
plt.ylabel('Calificación Promedio', fontsize=12)
plt.ylim(df_plot['Promedio General de Calificaciones'].min() * 0.9,
         df_plot['Promedio General de Calificaciones'].max() * 1.05) # Ajusta el límite Y para ver mejor la variación

# Opcional: Rotar etiquetas si los nombres de las tiendas son largos
plt.xticks(rotation=45, ha='right')

# 5. Mostrar el gráfico
plt.tight_layout()
plt.show()
```
