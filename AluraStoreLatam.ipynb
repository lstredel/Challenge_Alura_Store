#1. Análisis de facturación

```python
pd.options.display.float_format = '{:,.0f}'.format
```

```python
ventas_totales_por_tienda = Tiendas_Concatenadas.groupby("Tienda Origen")["Precio"].agg(["sum","count"])
ventas_totales_por_tienda.head()
```

# 2. Ventas por categoría

##2.1 Segmentacion de las ventas por Categoria del Producto

```python
ventas_totales_por_categoria = Tiendas_Concatenadas.groupby(["Categoría del Producto"])["Precio"].agg(["sum", "count"])
display(ventas_totales_por_categoria.head(10))
```

##2.2 Segmentacion de las Ventas Totales por Categoria y por *Tienda*

```python
ventas_totales_por_tienda_y_categoria = Tiendas_Concatenadas.groupby(["Tienda Origen", "Categoría del Producto"])["Precio"].agg(["sum", "count"])
display(ventas_totales_por_tienda_y_categoria.head(50))
```

# 3. Calificación promedio de la tienda


##3.1 Totalizacion y Segmentacion de las Calificaciones de Satisfaccion obtenidas por Tienda

```python
Calificacion_Tiendas = Tiendas_Concatenadas.groupby(["Tienda Origen", "Calificación"])["Calificación"].count().rename("Cantidad")
Calificacion_Tiendas.head(30)

```

##3.2 Totalizacion de Calificaciones Obtenidas

```python
total_por_tienda = Tiendas_Concatenadas.groupby("Tienda Origen")["Calificación"].count().rename("Total Calificado")
total_por_tienda.head()
```

##3.3 Creacion de df de frecuencias y totales para la obtencion de promedio

```python
df_frecuencias = Calificacion_Tiendas.reset_index()
```

```python
df_totales = total_por_tienda.reset_index()
```

##3.4 Creacion de columna Total Calificado y luego calculo de promedio

```python
df_combinado = df_frecuencias.merge(
    df_totales[['Tienda Origen', 'Total Calificado']],
    on='Tienda Origen',
    how='left'
)
df_combinado.head(30)
```

```python
df_combinado['Porcentaje por Calificación'] = (
    df_combinado['Cantidad'] / df_combinado['Total Calificado']
) * 100
pd.options.display.float_format = '{:,.0f}'.format
df_combinado.head(30)
```

# 4. Productos más y menos vendidos

##4.1 Creacion de df agrupado por producto y conteo de unidades vendidas para conocer el producto mas vendido

```python
df_ventas_por_producto = Tiendas_Concatenadas.groupby('Producto').size().reset_index(name='Total_Cantidad_Vendida_Global')
```

##4.2 Creacion de df agrupado por producto y conteo de unidades vendidas para conocer el TopTen de producto mas vendido

```python
N = 10
df_productos_top_10 = df_ventas_por_producto.sort_values(
    by='Total_Cantidad_Vendida_Global',  # Columna de conteo
    ascending=False                     # De mayor a menor
).head(N)
print(f"--- Top {N} Productos Más Vendidos (por Conteo) ---")
display(df_productos_top_10)
```

##4.3 Creacion de df agrupado por producto y conteo de unidades vendidas para conocer el TopBottom de producto menos vendido

```python
N = 10
df_productos_bottom_10 = df_ventas_por_producto.sort_values(
    by='Total_Cantidad_Vendida_Global',  # Columna de conteo
    ascending=True                     # De menor a mayor
).head(N)
print(f"--- Top {N} Productos Menos Vendidos (por Conteo) ---")
display(df_productos_bottom_10)
```

##4.4 Segmentacion y Participacion por tienda de los Productos mas Vendidos

```python
productos_top_global = df_productos_top_10['Producto'].unique()
df_solo_top_productos = Tiendas_Concatenadas[
    Tiendas_Concatenadas['Producto'].isin(productos_top_global)
]
df_participacion_tienda = df_solo_top_productos.groupby(['Producto', 'Tienda Origen']).size().reset_index(name='Cantidad_Vendida_Tienda')
df_participacion_final = df_participacion_tienda.merge(
    df_ventas_por_producto,
    on='Producto',
    how='left'
)
df_participacion_final['Porcentaje_Participacion'] = (
    df_participacion_final['Cantidad_Vendida_Tienda'] / df_participacion_final['Total_Cantidad_Vendida_Global']
) * 100
pd.options.display.float_format = '{:,.0f}'.format

resumen_styled = df_participacion_final.style.format({
    'Cantidad_Vendida_Tienda': '{:,.0f}',
    'Total_Cantidad_Vendida_Global': '{:,.0f}',
    'Porcentaje_Participacion': '{:,.2f}%' # Porcentaje con 2 decimales
})

print("--- Porcentaje de Participación por Tienda (Solo Productos Top 10) ---")
display(resumen_styled)
```

#7. Localizacion de Ventas

##7.1 Importacion de Biblioteca Folium para la geolocalizacion

```python
import pandas as pd
import folium
from folium.plugins import HeatMap
```

```python
lat_centro = Tiendas_Concatenadas['lat'].mean()
lon_centro = Tiendas_Concatenadas['lon'].mean()

# 3. Crear el mapa base
# zoom_start=10 es un buen punto de inicio para una ciudad o región
mapa_ventas = folium.Map(location=[lat_centro, lon_centro], zoom_start=10)

# Preparar los datos para el heatmap
datos_heatmap = Tiendas_Concatenadas[['lat', 'lon']].values.tolist()

# 4. Añadir la capa de Mapa de Calor
HeatMap(datos_heatmap).add_to(mapa_ventas)

# 5. Guardar o mostrar el mapa (se guarda como archivo HTML)
mapa_ventas.save("mapa_ventas_interactivo.html")

print("✅ Mapa interactivo guardado como 'mapa_ventas_interactivo.html'. ¡Ábrelo en tu navegador!")
```

```python
df = Tiendas_Concatenadas.copy()
lat_centro = df['lat'].mean()
lon_centro = df['lon'].mean()

# Obtener una lista de tiendas únicas
tientas_unicas = df['Tienda Origen'].unique()

# 2. Crear el mapa base (usamos zoom_start=10 como ejemplo)
mapa_tiendas = folium.Map(location=[lat_centro, lon_centro], zoom_start=10)

# 3. Iterar sobre cada tienda y crear una capa de HeatMap separada
for tienda in tientas_unicas:
    # 3a. Filtrar los datos para la tienda actual
    df_tienda = df[df['Tienda Origen'] == tienda]

    # Preparamos los datos para el HeatMap: [lat, lon, peso (Precio)]
    # Si quieres usar el conteo de transacciones, cambia 'Precio' por un valor constante como 1
    datos_heatmap = df_tienda[['lat', 'lon', 'Precio']].values.tolist()

    # 3b. Crear un grupo de características (FeatureGroup) para la tienda
    # Esto permite que la capa sea controlable en el mapa
    fg = folium.FeatureGroup(name=f'Ventas - {tienda}')

    # 3c. Añadir el HeatMap a ese FeatureGroup
    HeatMap(datos_heatmap).add_to(fg)

    # 3d. Añadir el FeatureGroup al mapa
    fg.add_to(mapa_tiendas)

# 4. Añadir el control de capas
# Esto crea los botones en la esquina superior derecha del mapa
from folium import LayerControl
LayerControl().add_to(mapa_tiendas)

# 5. Guardar o mostrar el mapa
nombre_archivo = "mapa_ventas_por_tienda_interactivo.html"
mapa_tiendas.save(nombre_archivo)

print(f"✅ Mapa con capas guardado como '{nombre_archivo}'.")
print("Descárgalo y ábrelo. Usa el icono de capas (☰) en la esquina superior derecha para activar/desactivar cada tienda.")
```
