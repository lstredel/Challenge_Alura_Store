### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
```

```python
tienda1.insert(loc=0, column='Tienda Origen', value="Tienda 1")
tienda2.insert(loc=0, column='Tienda Origen', value="Tienda 2")
tienda3.insert(loc=0, column='Tienda Origen', value="Tienda 3")
tienda4.insert(loc=0, column='Tienda Origen', value="Tienda 4")
```

```python
Tiendas_Concatenadas = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)
```

#1. Análisis de facturación

```python
ventas_totales_por_tienda = Tiendas_Concatenadas.groupby("Tienda Origen")["Precio"].sum()
ventas_totales_por_tienda.head()
```

# 2. Ventas por categoría

```python
ventas_totales_por_categoria = Tiendas_Concatenadas.groupby(["Categoría del Producto"])["Precio"].agg(["sum", "count"])
display(ventas_totales_por_categoria.head(10))
```

```python
ventas_totales_por_tienda_y_categoria = Tiendas_Concatenadas.groupby(["Tienda Origen", "Categoría del Producto"])["Precio"].agg(["sum", "count"])
display(ventas_totales_por_tienda_y_categoria.head(50))
```

# 3. Calificación promedio de la tienda


```python
Calificacion_Tiendas = Tiendas_Concatenadas.groupby(["Tienda Origen", "Calificación"])["Calificación"].count().rename("Cantidad")
Calificacion_Tiendas.head(30)

```

```python
total_por_tienda = Tiendas_Concatenadas.groupby("Tienda Origen")["Calificación"].count().rename("Total Calificado")
total_por_tienda.head()
```

```python
df_frecuencias = Calificacion_Tiendas.reset_index()
```

```python
df_totales = total_por_tienda.reset_index()
```

```python
df_combinado = df_frecuencias.merge(
    df_totales[['Tienda Origen', 'Total Calificado']],
    on='Tienda Origen',
    how='left'
)
df_combinado.head(30)
```

```python
df_combinado['Porcentaje por Calificación'] = (
    df_combinado['Cantidad'] / df_combinado['Total Calificado']
) * 100
pd.options.display.float_format = '{:,.0f}'.format
df_combinado.head(30)
```

# 4. Productos más y menos vendidos

```python
ventas_totales_categoria = Tiendas_Concatenadas.groupby(["Categoría del Producto"])["Precio"].agg(["sum", "count"])
ordenado_descendente = ventas_totales_por_tienda_y_categoria.sort_values(by="sum", ascending=False)
df_venta_productos = df_venta_productos.reset_index()
display(ordenado_descendente.head(10))
```

```python
ventas_totales_por_tienda_y_categoria = Tiendas_Concatenadas.groupby(["Categoría del Producto", "Tienda Origen"])["Precio"].agg(["sum"])
df_venta_productos = ventas_totales_por_tienda_y_categoria.rename(columns={'sum': 'Ventas_Tienda'})
ventas_totales_categoria = Tiendas_Concatenadas.groupby(["Categoría del Producto"])["Precio"].agg(["sum"])
df_totales_categoria = ventas_totales_categoria.rename(
    columns={'sum': 'Ventas_Total_Global_Categoria'}
)
df_venta_productos = df_venta_productos.merge(
    df_totales_categoria[['Ventas_Total_Global_Categoria']],
    on='Categoría del Producto',
    how='left'  # Mantiene todas las filas del DF de detalle
)
df_venta_productos['Porcentaje_Venta_Tienda'] = (
    df_venta_productos['Ventas_Tienda'] / df_venta_productos['Ventas_Total_Global_Categoria']
) * 100
display(df_venta_productos.head(50))
```
