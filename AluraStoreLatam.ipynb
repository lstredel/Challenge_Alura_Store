### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)
```

```python
tienda1.insert(loc=0, column='Tienda Origen', value="Tienda 1")
tienda2.insert(loc=0, column='Tienda Origen', value="Tienda 2")
tienda3.insert(loc=0, column='Tienda Origen', value="Tienda 3")
tienda4.insert(loc=0, column='Tienda Origen', value="Tienda 4")
```

```python
Tiendas_Concatenadas = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)
```

#1. Análisis de facturación



```python
pd.options.display.float_format = '{:,.0f}'.format
```

```python
ventas_totales_por_tienda = Tiendas_Concatenadas.groupby("Tienda Origen")["Precio"].sum()
ventas_totales_por_tienda.head()
```

# 2. Ventas por categoría

```python
ventas_totales_por_categoria = Tiendas_Concatenadas.groupby(["Categoría del Producto"])["Precio"].agg(["sum", "count"])
display(ventas_totales_por_categoria.head(10))
```

```python
ventas_totales_por_tienda_y_categoria = Tiendas_Concatenadas.groupby(["Tienda Origen", "Categoría del Producto"])["Precio"].agg(["sum", "count"])
display(ventas_totales_por_tienda_y_categoria.head(50))
```

# 3. Calificación promedio de la tienda


```python
Calificacion_Tiendas = Tiendas_Concatenadas.groupby(["Tienda Origen", "Calificación"])["Calificación"].count().rename("Cantidad")
total_por_tienda = Tiendas_Concatenadas.groupby("Tienda Origen")["Calificación"].count().rename("Total Calificado")

```

```python
df_frecuencias = Calificacion_Tiendas.reset_index()
```

```python
df_frecuencias = Calificacion_Tiendas.reset_index()
df_frecuencias = df_frecuencias.merge(total_por_tienda, on="Tienda Origen")
df_frecuencias["Frecuencia Relativa"] = df_frecuencias["Cantidad"] / df_frecuencias["Total Calificado"]

print("Frecuencia de calificaciones por tienda:")
display(df_frecuencias.head())

print("\nCalificación promedio por tienda:")
calificacion_promedio_por_tienda = Tiendas_Concatenadas.groupby("Tienda Origen")["Calificación"].mean()
display(calificacion_promedio_por_tienda.head())
```
